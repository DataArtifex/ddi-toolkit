```python
from dartfx.ddi.ddicodebook import loadxml
from dartfx.ddi.utils import codebook_to_cdif
from dartfx.rdf.utils import get_rdf_graph_statistics
import json
import logging
logging.basicConfig(level=logging.DEBUG)
```

```python
# load codebook
cb = loadxml('../tests/ddi/codebook/simple_yndk.xml')
#cb = loadxml('../tests/ddi/codebook/NES1948.xml')
#cb = loadxml('../tests/ddi/codebook/socrata_sfo_311.xml')
```

```python
# convert DDI-C to DDI-CDIF resources
resources = codebook_to_cdif(cb)
json_data = ""
for resource in resources.values():
    json_data += resource.as_json(indent=4) + ",\n"
with open('lab_ddic2cdi.json', 'w') as outfile:
    outfile.write("[\n" +json_data[:-2] + "\n]")
```

```python
# generate and save Graph
from rdflib import Graph
g = Graph()
#list(resources.values())[0].get_uriref()
for resource in resources.values():
    resource.add_to_rdf_graph(g)
g.serialize(destination='lab_ddic2cdi.ttl', format="turtle")
g.serialize(destination='lab_ddic2cdi.jsonld', format="json-ld", indent=4)
g.serialize(destination='lab_ddic2cdi.nt', format="nt")

with open('lab_ddic2cdi.stats.json', 'w') as outfile:
    json.dump(get_rdf_graph_statistics(g), outfile, indent=4)
```
