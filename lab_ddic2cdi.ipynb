```python
from dartfx.ddi.ddicdi import InstanceVariable, ControlledVocabularyEntry, VariableStructure
from dartfx.ddi.ddicodebook import loadxml
from dartfx.ddi.utils import codebook_to_cdi
import logging
import typing
import uuid
logging.basicConfig(level=logging.DEBUG)
```

```python
foo = InstanceVariable()
foo.set_simple_name("V1")
print(f"name\n{foo.name[0]}\n")
foo.set_simple_display_label("Variable1")
print(f"displayLabel\n{foo.displayLabel[0].languageSpecificString[0]}\n")
foo.set_simple_identifier_ddi(str(uuid.uuid4()))
print(f"identifier:ddiIdentifier\n{foo.identifier.ddiIdentifier}\n")
foo.set_simple_identifier_nonddi(str(uuid.uuid4()))
print(f"identifier:nonddiIdentifier\n{foo.identifier.nonDdiIdentifier}\n")
foo.set_simple_identifier_uri(str(f"urn:uuid:{str(uuid.uuid4())}"))
print(f"identifier:uri\n{foo.identifier.uri}\n")

print("FOO\n")
print(foo)

foo.save_json("foo.json")
```
