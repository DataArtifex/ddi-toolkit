```python
from dartfx.ddi.ddicdi import InstanceVariable, ControlledVocabularyEntry, VariableStructure, RepresentedVariable
from dartfx.ddi.ddicodebook import loadxml
from dartfx.ddi.utils import codebook_to_cdif, get_rdf_graph_statistics
import logging
logging.basicConfig(level=logging.DEBUG)
```

```python
cb = loadxml('tests/ddi/codebook/simple_yndk.xml')
#cb = loadxml('tests/ddi/codebook/NES1948.xml')
```

```python
resources = codebook_to_cdif(cb)
json_data = ""
for resource in resources.values():
    json_data += resource.as_json(indent=4) + ",\n"
with open('lab_ddic2cdi.json', 'w') as outfile:
    outfile.write("[\n" +json_data[:-2] + "\n]")
```

```python
from rdflib import Graph
g = Graph()
#list(resources.values())[0].get_uriref()
for resource in resources.values():
    resource.add_to_rdf_graph(g)
g.serialize(destination='lab_ddic2cdi.ttl', format="turtle")
```

```python
get_rdf_graph_statistics(g)
```

```python
query = """
SELECT ?type ?property (COUNT(?property) AS ?n)
WHERE {
  ?resource a ?type .
  ?resource ?property ?value .
}
GROUP BY ?type ?property
ORDER BY ?type ?property
"""
for row in g.query(query):
    pass
    print(str(g.namespace_manager.qname(row.type))+' '+str(g.namespace_manager.qname(row.property))+' '+str(row.n))
```
